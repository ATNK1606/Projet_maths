```python
#Priorité
#Change le nom de ce fihier d'abord
#Run ce fichier.
import numpy as np
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.datasets import cifar10
import pandas as pd

# =========================
# 1. Chargement des données
# =========================
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# Sélection de deux classes (ex: avion=0, voiture=1)
classes = [0, 1]

train_mask = np.isin(y_train, classes).flatten()
test_mask = np.isin(y_test, classes).flatten()

X_train, y_train = X_train[train_mask], y_train[train_mask]
X_test, y_test = X_test[test_mask], y_test[test_mask]

# Re-normalisation
X_train = X_train.astype("float32") / 255.0
X_test = X_test.astype("float32") / 255.0

# Re-labeling (0 et 1)
y_train = (y_train == classes[1]).astype(int)
y_test = (y_test == classes[1]).astype(int)

y_train = to_categorical(y_train, 2)
y_test = to_categorical(y_test, 2)

# =========================
# 2. Construction du CNN
# =========================
model = Sequential()

model.add(Conv2D(32, (3,3), activation='relu', padding='same', input_shape=(32,32,3)))
model.add(Conv2D(32, (3,3), activation='relu'))
model.add(MaxPooling2D((2,2)))

model.add(Conv2D(64, (3,3), activation='relu', padding='same'))
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D((2,2)))

model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(2, activation='softmax'))

# =========================
# 3. Compilation
# =========================
model.compile(
    optimizer=Adam(learning_rate=1e-3),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# =========================
# 4. Entraînement
# =========================
epochs = 5
batch_size = 32

history = model.fit(
    X_train, y_train,
    epochs=epochs,
    batch_size=batch_size,
    validation_split=0.2
)

# =========================
# 5. Tableau des epochs
# =========================
df_epochs = pd.DataFrame({
    "Epoch": range(1, epochs + 1),
    "Loss": history.history["loss"],
    "Accuracy": history.history["accuracy"],
    "Val Loss": history.history["val_loss"],
    "Val Accuracy": history.history["val_accuracy"]
})

print("\n=== Résultats par epoch ===")
print(df_epochs)

# =========================
# 6. Évaluation finale (TEST)
# =========================
test_loss, test_accuracy = model.evaluate(X_test, y_test)

print("\n=== Résultat FINAL ===")
print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)

# =========================
# 7. Sauvegarde NUMPY
# =========================
results = {
    "epoch_table": df_epochs.to_dict(orient="list"),
    "final_train_accuracy": history.history["accuracy"][-1],
    "final_val_accuracy": history.history["val_accuracy"][-1],
    "test_accuracy": test_accuracy,
    "test_loss": test_loss
}

np.save("IngenierieMath_DM_2025_KEITA_Amadou_Tamba_Namory_essai.npy", results)

print("\nFichier .npy sauvegardé avec succès.")
```
